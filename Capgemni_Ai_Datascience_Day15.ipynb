```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
from sklearn.metrics import accuracy_score
import numpy as np

data = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/refs/heads/master/adult-train.csv')
data.head()
```

```python
data[' Bachelors'].unique()
```

```python
data[' State-gov'].unique()
```

```python
data.replace(' ?',np.nan,inplace=True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
lb = LabelEncoder()
data[' State-gov']=lb.fit_transform(data[' State-gov'])
data[' Never-married']=lb.fit_transform(data[' Never-married'])
data[' Adm-clerical']=lb.fit_transform(data[' Adm-clerical'])
data[' Not-in-family']=lb.fit_transform(data[' Not-in-family'])
data[' White']=lb.fit_transform(data[' White'])
data[' Male']=lb.fit_transform(data[' Male'])
data[' United-States']=lb.fit_transform(data[' United-States'])
data[' Bachelors']=lb.fit_transform(data[[' Bachelors']])
```

```python
data.head()
```

```python
x=data.drop(' <=50K',axis=1)
y=data[' <=50K']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier(random_state=1)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=1)
model.fit(x_train,y_train)
y_pred2=model.predict(x_test)
y_pred2
```

```python
accuracy_score(y_test,y_pred2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=1)
model.fit(x_train,y_train)
y_pred3=model.predict(x_test)
y_pred3
```

```python
accuracy_score(y_test,y_pred3)
```

```python
from sklearn.ensemble import BaggingClassifier
model = BaggingClassifier(random_state=1)
model.fit(x_train,y_train)
y_pred4=model.predict(x_test)
y_pred4
```

```python
accuracy_score(y_test,y_pred4)
```
