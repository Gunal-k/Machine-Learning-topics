```python
#feature Selection
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest,chi2

df=load_iris()
x=df.data
y=df.target

#feature selection using chi2 scoring function
num_feature = 2
feature_select = SelectKBest(score_func = chi2, k = num_feature)
x_selected = feature_select.fit_transform(x, y)

#print selected features
features = feature_select.get_support(indices = True)
features
```

```python
#feature Extraction
from sklearn.decomposition import PCA
df = load_iris()
x = df.data
#feature extraction using PCA
num_components = 1
pca = PCA(n_components = num_components)
x_transform = pca.fit_transform(x)
#display result
print("original: ", x.shape)
print('transformed: ', x_transform.shape)
```

```python
#regression
df = pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/Salary_Data.csv')
df
```

```python
#take input and output features
x = df.iloc[:, 0:1]
y = df.iloc[:, 1]
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
y_pred
```

```python
plt.scatter(x, y, c= 'g')
plt.plot(x, y_pred, color = 'pink')
plt.title('Best fit line')
plt.xlabel('position')
plt.ylabel('salary')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv("https://raw.githubusercontent.com/pcsanwald/kaggle-titanic/master/train.csv")
df
```

```python
pass_with_cabin = df[pd.notnull(df['cabin'])]
pass_with_cabin['deck'] = pass_with_cabin['cabin'].str[0]
survival_rate_by_deck = (
  pass_with_cabin.groupby('deck')
  .agg(survival_rate=('survived','mean'))
  .reset_index()
  .sort_values(by='survival_rate',ascending=False)
)
plt.pie(
  survival_rate_by_deck['survival_rate'],_
  labels=survival_rate_by_deck['deck'],
  autopct='%1.1f%%'
)
plt.title('survival rate by deck')
plt.show()
```
