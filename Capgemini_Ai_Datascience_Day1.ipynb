```python
#function to calculate tax based on their income
def taxCalc(income):
  tax=0
  if income<300000:
   return tax
  elif income in range(300000,500001):
    return income*0.1
  elif income in range(500000,1000001):
    return income*0.12
  elif income in range(1000000,2000001):
    return income*0.13
  else:
    return income*0.134

#taking input from user
income = int(input("Enter your Income : "))
print("The Tax calculated is ",taxCalc(income))

```
