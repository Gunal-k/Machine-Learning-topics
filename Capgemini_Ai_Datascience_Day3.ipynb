```python
#function to cipher the given message
def cipher(msg):
  encrypted_message=''
  for i in msg:
    asci_val=ord(i)
    if asci_val>0:
      encrypted_message+=chr(asci_val+1)
    else:
      encrypted_message+=i
  return encrypted_message

#taking input from user
message = input("Enter the String : ")
print(cipher(message))
```
