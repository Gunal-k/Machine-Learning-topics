```python
#program to multiply matrices, takes input from users and prints it
import random
def matrix_multiplication():
  col=int(input("Enter the number of columns : "))
  row=int(input("Enter the number of rows : "))
  a = [[random.randrange(1, 10) for _ in range(col)] for _ in range(row)]
  b = [[random.randrange(1, 10) for _ in range(col)] for _ in range(row)]
  c=[[0 for _ in range(col)] for _ in range(row)]

  for i in range(row):
    for j in range(col):
      for k in range(col):
        c[i][j]+=a[i][k]*b[k][j]

  print("Matrix A")
  for i in a:
    print(i)
  print("\nMatrix B")
  for i in b:
    print(i)

  print("\nMultiplied Matrix Value")
  for i in c:
    print(i)

matrix_multiplication()
```

```python
import random
#program for guessing a number generated randomly by computer using hint (higher/lower)
guessed_num=random.randrange(1,9)
number=int(input("Guess the Number : "))
while True:
  if number==guessed_num:
    break
  if(number>guessed_num):
    print("Your Guess is Higher")
  else:
    print("Your Guess is Lower")
  number=int(input("Predict the Magical Number : "))
print("You Won, Yay")
```

```python
#function to find factorial of a given number
def fact(num):
  if num==0:
    return 1
  return num*fact(num-1)

#Taking input from user
number = int(input("Enter the Number : "))
print("Factorial of",number,"is",fact(number))
```

```python
#function to check a string is palindrome or not
def palindrome(msg):
  for i in range(len(msg)//2):
    if msg[i]==msg[len(msg)-i-1]:
      continue
    else:
      return False
  return True

#taking input from user
message = input("Enter the String : ")
if palindrome(message):
  print(message+" is a Palindrome")
else:
  print(message+" is not a Palindrome")
```

```python
#function to check the number is prime or not
def prime(num):
  count=0
  for i in range(2,num//2+1):
    if(num%i==0):
      count+=1
    if(count==1):
      return False
  return True

#taking input from user
number = int(input("Enter the Number : "))
if prime(number):
  print(number,"is a Prime number")
else:
  print(number," is not a Prime number")
```

```python
import random
#function to sum a list of numbers
def sum(num):
  sum_num=0
  for i in num:
    sum_num+=i
  return sum_num

#taking input from user
number = int(input("Enter the Number of values : "))
num_list=[int(input("Enter the value : ")) for _ in range(number)]
print(num_list)
sum=sum(num_list)
print("\nSum of the List = ",sum)
```

```python
#function to print fibonacci series
def fib(num):
  num1=0
  num2=1
  while True:
    if num==0:
      break
    fib_val=num1+num2
    num1=num2
    num2=fib_val
    print(fib_val,end=" ")
    num-=1

#taking input from user
number = int(input("Enter the Limit to display Fibonacci series : "))
fib(number)
```

```python
#function to reverse a given number
def reverse(num):
  rev_num=0
  while True:
    if num==0:
      break
    a=num%10
    rev_num=rev_num*10+a
    num//=10
  return rev_num

#taking input from user
number = int(input("Enter the Number : "))
rev_num=reverse(number)
print("Reverse of number",number,"=",rev_num)
```

```python
#Function to find the given number is odd or even
def odd_even(num):
  if num%2==0:
    return True
  else:
    return False

#taking input from user
number = int(input("Enter the Number : "))
if odd_even(number):
  print(number,"is Even")
else:
  print(number,"is Odd")
```

```python
import math
#Function to find square root of a given number
def root(num):
  return round(math.sqrt(num))

#taking input from user
number = int(input("Enter the Number : "))
print("The Square of",number,"is",root(number))

```

```python
#Function to find power of a given number
def power(base,exponent):
  return base**exponent

#taking input from user
base = int(input("Enter the Base : "))
exponent = int(input("Enter the Exponent : "))
print(base,"raised to",exponent,"=",power(base,exponent))
```

```python
#Function to concatenate given Strings
def concat(str1,str2):
  return str1+" "+str2

#taking input from user
str1 = input("Enter the String 1 : ")
str2 = input("Enter the String 2 : ")
print("Before Concatenation")
print(str1)
print(str2)
print("\nAfter Concatenation")
print(concat(str1,str2))
```
