```python
import pandas as pd

# load the data from the URL
url = "https://raw.githubusercontent.com/sudarshan-koirala/Salary-Prediciton-based-on-Years-of-Experience/master/Salary_Data.csv"
data = pd.read_csv(url)

#display the first few rows of the dataset
print("First few rows of the dataset : ")
print(data.head())
```

```python
# Display basic information about the dataset
print("\nBasic information about the dataset:")
print(data.info())
```

```python
# Display summary statistics of numerical columns
print("\nSummary statistics of numerical columns:")
print(data.describe())
```

```python
# Display the shape of the dataset (number of rows and colums)
print("\nShape of the dataset:")
print(data.shape)
```

```python
# Display datatypes of each column
print("\nData types of each column:")
print(data.dtypes)
```

```python
#checking for missing values
print("\nMissing values in the dataset")
print(data.isnull().sum())
```

```python
#checking for duplicate values
print("\nMissing values in the dataset")
print(data.duplicated().sum())
```

```python
#remove duplicates if any
data = data.drop_duplicates()
```

```python
#rename columns if needed
data = data.rename(columns={"YearsExperience":"Exp"})
```

```python
#convert data types if needed
data["Exp"]=data["Exp"].astype(int)
data.head()
```

```python
#Selecting a single column
print("\nSelecting a single column : ")
print(data["Exp"])
```

```python
#Selecting multiple colums
print("\nSelecting a single column : ")
print(data[["Exp","Salary"]])
```

```python
# Filtering rows based on a condition
print("\nFiltering rows based on a condition:")
print(data[data["Exp"] <5])
```

```python
#Adding new column
data["Salary_in_thousands"] = data["Salary"]/1000
data["Exp_in_months"] = data["Exp"] * 12
data.head()
```

```python
#Grouping data by a column and calculating statistics
#calculating the average salary for each years of experience
grouped_data = data.groupby("Exp_in_months").mean()
print("\ngrouped data - Average salary for each years of experience: ")
print(grouped_data)
```

```python
#reshaping data (pivot table, melt, stack, unstack)
#pivot table to calculate the average salary for each years of experience
pivot_table=data.pivot_table(index="Exp",values="Salary_in_thousands",aggfunc="mean")
print(pivot_table)
```
